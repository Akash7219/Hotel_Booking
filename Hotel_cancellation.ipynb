```python
def adult_and_child(data):
    if data['adults']==1 and (data['children']==0 and data['babies']==0):
        return 'single_only'
    elif data['adults']>0 and (data['children']>0 or data['babies']>0):
        return 'with_family'
    elif data['adults']<3 and (data['children']==0 and data['babies']==0):
        return 'small_group'
    elif 3<=data['adults']<=6 and (data['children']==0 and data['babies']==0):
        return 'medium_group'
    elif data['adults']>6 and (data['children']==0 and data['babies']==0):
        return 'large_group'
data['with_family_or_else']=data.apply(adult_and_child,axis=1)
```

```python
#Creating new column on the basis of the number of nights spent.
def day_night_count(data):
    if data['total_nights_stayed']<7:
        if data['stays_in_weekend_nights']>0 and data['stays_in_week_nights']==0:
            return 'weekend_stay_only'
        elif data['stays_in_weekend_nights']==0 and data['stays_in_week_nights']>0:
            return 'week_stay_only'
        elif data['stays_in_weekend_nights']==0 and data['stays_in_week_nights']==0:
            return 'no_stay'
        elif data['stays_in_weekend_nights']>0 and data['stays_in_week_nights']>0:
            return 'weekday_and_weekend_stay'
    elif data['total_nights_stayed']==7:
        return 'week_long_stay'
    else:
        return 'long_stay'
data['type_of_stay']=data.apply(day_night_count,axis=1)
data[['type_of_stay','stays_in_weekend_nights','stays_in_week_nights']].head(5)
```
